Ploting recorded data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import csv

x = []
y = []

with open('tacx_sirius_5_no_header.cvs','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(float(row[2]))
        y.append(float(row[3]))

plt.plot(x,y, 'x')
plt.xlabel('Speed, km/h')
plt.ylabel('Power, W')
plt.title('Taxc Sirius power curve')
#plt.legend()
plt.show()
```

As it is seen from the graph all points lie on a straight line.

Parameters of the line can be obtained by the method of least squares.

y = m*x + c

```python
X = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(X, y)[0]
print m, c
```

Plot the data along with the fitted line:

```python
x= np.asarray(x)
plt.plot(x, y, 'x', label='Original data')
plt.plot(x, m*x + c, 'r', label='Fitted line')
plt.xlabel('Speed, km/h')
plt.ylabel('Power, W')
plt.title('Taxc Sirius power curve')
plt.legend()
plt.show()
```
